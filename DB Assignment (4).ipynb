```python
from pyspark.sql.functions import *
df1 = df.withColumn("Full Name", concat_ws(" ", "FirstName", "MiddleName", "LastName"))
```

```python
df2 = df1.withColumn("Contact",regexp_replace("Phone", r"[- ()]", "").substr(-10, 10).cast('long'))
```

```python
df3 = df2.drop("Suffix", "Phone", "FirstName", "MiddleName", "LastName")
```

```python
df4 = df3.withColumn("SalesPerson", regexp_replace("SalesPerson", r"adventure-works\\\\", ""))
```
