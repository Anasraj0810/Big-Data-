```python
df = spark.read.csv('dbfs:/Volumes/workspace/default/anas_dataset/CustomerSourceDemo.csv', header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import *

df.groupBy("Gender").agg(
sum("Salary").alias("Total Salary")
,max("Salary").alias("Max Salary"),
min("Salary").alias("Min Salary"),
avg("Salary").alias("Avg Salary")).show()
```
